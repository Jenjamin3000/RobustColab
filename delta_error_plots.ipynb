```python
from matplotlib import pyplot as plt
from Plots import plot_multiple_epochs
from Utils import *
DATA = './data/delta_error_plots/'
```

```python

N_USERS = 1000
N_ITEMS = 1000
MISSING_PROB = 0.8
NB_ATTACKERS = np.arange(0, 4, 3)
N_EPOCH = 1
```

```python
ax = plot_multiple_epochs(n_users=N_USERS, n_items=N_ITEMS, missing_prob=MISSING_PROB, nb_attackers_to_test=NB_ATTACKERS,
                          error_func='delta', nb_epochs=N_EPOCH, attackers_algo='random')
plt.savefig(DATA+'delta_random.png')
plt.show()

ax = plot_multiple_epochs(n_users=N_USERS, n_items=N_ITEMS, missing_prob=MISSING_PROB, nb_attackers_to_test=NB_ATTACKERS,
                          error_func='delta', nb_epochs=N_EPOCH, attackers_algo='targeted')
plt.savefig(DATA+'delta_targeted.png')
plt.show()

ax = plot_multiple_epochs(n_users=N_USERS, n_items=N_ITEMS, missing_prob=MISSING_PROB, nb_attackers_to_test=NB_ATTACKERS,
                          error_func='error', nb_epochs=N_EPOCH, attackers_algo='random')
plt.savefig(DATA+'error_random.png')
plt.show()
```
